```python
for keys in us_state_abbrev.keys():
    us_state_abbrev[keys] = "us-" + us_state_abbrev[keys].lower()
```

```python
state_temps = {}
curr = None
first_records = datetime.strptime("1992 01 01", "%Y %m %d")

for i in range(len(df1)):
    curr = df1.iloc[i]
    if us_state_abbrev[curr.name[0]] in state_temps:
        if curr.name[1] > 1992: 
            state_temps[us_state_abbrev[curr.name[0]]]["dates"].append(int(curr.name[1]))
            state_temps[us_state_abbrev[curr.name[0]]]["avg_temp"].append(float(curr.AverageTemperature.round(3)))
            state_temps[us_state_abbrev[curr.name[0]]]["avg_uncertain"].append(float(curr.AverageTemperatureUncertainty))
    else:
        state_temps[us_state_abbrev[curr.name[0]]] = {"dates": [], "avg_temp": [], "avg_uncertain": []}
```
